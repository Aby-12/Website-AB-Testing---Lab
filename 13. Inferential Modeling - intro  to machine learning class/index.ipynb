* Independent variable -- the input(s) you think influence Y (predictors/features).

```python
import pandas as pd

salary_df=pd.read_csv('salary_data.csv')
salary_df.head()
```

```python
# plot a scatter plot between x (years of experince-independent variable) and y (salary)

import matplotlib.pyplot as plt
plt.scatter (salary_df['YearsExperience'], salary_df['Salary'])
plt.title('Scatter plot for salary against years of experience')
plt.ylabel('Y')
plt.xlabel('X')
plt.show()

##There is a linear relationship between the two variables (positive one--as year of experience increases, salary increases)

# creat a line of best fit.. we create least squares estimation
```

Least squares estimation is a method for fitting a line (or regression model) by choosing the coefficients that make the model’s predictions as close as possible to the actual data.

“Close as possible” is defined as: minimize the sum of squared errors (residuals):

So it finds the line where the total squared vertical distances from the points to the line is smallest.
