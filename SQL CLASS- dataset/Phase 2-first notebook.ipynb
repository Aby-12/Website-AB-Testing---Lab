```python
import sqlite3

# create or open database
conn = sqlite3.connect("my_database.sqlite")
cur = conn.cursor()

# create table
cur.execute("""
CREATE TABLE IF NOT EXISTS people (
  id INTEGER PRIMARY KEY,
  name TEXT,
  age INTEGER,
  email TEXT
);
""")

# insert sample data
cur.execute("INSERT INTO people (name, age, email) VALUES ('Bob', 29, 'bob@flatironschool.com');")
cur.execute("INSERT INTO people (name, age, email) VALUES ('James', 28, 'james@flatironschool.com');")
conn.commit()

# show tables
cur.execute("SELECT name FROM sqlite_master WHERE type='table';")
print("Tables:", cur.fetchall())

# show rows
cur.execute("SELECT * FROM people;")
rows = cur.fetchall()
for r in rows:
    print(r)

conn.close()
```
